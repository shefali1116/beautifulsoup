Write a python program to scrape cricket rankings from 'www.icc-cricket.com'. you have to scrape:

i) Top 10 ODI teams in men's cricket along with the records for matches,points and ratings.

ii) Top 10 ODI Batsman in men along with the records of their team and rating.

iii) Top 10 ODI bowlers along with the records of their team and rating.

```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

lets see show the webpage source code downloaded above looks like

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
#lets extract all the tags having TOP10 ODI teams in men's cricket along with the records for matches,points and rating.
records=soup.find_all('table',class_="table")
records[0:10]
```

```python
#we have all the tags in which there are TOP 10 ODI teams in men's cricket along with the records for matches
#we will extract the text from these tags one by one by Looping over these tags

all_records=[]  #empty list for store the
for i in records[0:10]:
    all_records.append(i.text.replace('\n',''))
all_records
```

Now we will extract the Top 10 ODI Batsman in men along with the records of their team and rating

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
#we will find the Top 10 ODI Batsman in men along with the records of their team and rating
batsman=soup.find_all('table',class_="table rankings-table")
batsman[0:10]
```

```python
#we have all the tags in which there are Top 10 ODI Batsman in men along with the records of their team and rating
#we will extract the text from these tags one by one by Looping over these tags

all_batsman=[]  #empty list for store the
for i in batsman[0:10]:
    all_batsman.append(i.text.replace('\n',''))
all_batsman
```

 Top 10 ODI bowlers along with the records of their team and rating

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
#we will find the Top 10 ODI bowlers along with the records of their team and rating
bowlers=soup.find_all('tr',class_="table-body")
bowlers[0:10]
```

```python
#we have all the tags in which there are Top 10 ODI bowlers along with the records of their team and rating
#we will extract the text from these tags one by one by Looping over these tags

all_bowlers=[]  #empty list for store the
for i in bowlers[0:10]:
    all_bowlers.append(i.text.replace('\n',''))
all_bowlers
```
