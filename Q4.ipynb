```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/")
```

lets see show the webpage source code downloaded above looks like

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
#lets extract all the tags having the 5 book names
name=soup.find_all('h4',class_="italic")
name[0:5]
```

```python
#we have all the tags in which there are book names
#we will extract the text from these tags one by one by Looping over these tags

all_names=[]  #empty list for store the
for i in name[0:5]:
    all_names.append(i.text.replace('\n',''))
all_names
```

```python
#lets extract all the tags having the 5 book's Author names
author_name=soup.find_all('p',class_="sans bold")
author_name[0:5]
```

```python
#we have all the tags in which there are the 5 book names
#we will extract the text from these tags one by one by Looping over these tags

all_author=[]  #empty list for store the
for i in author_name[0:5]:
    all_author.append(i.text.replace('\n',''))
all_author
```

```python
#lets extract all the tags having the 5 books genre
genre=soup.find_all('p',class_="genre-links hidden-phone")
genre[0:5]
```

```python
#we have all the tags in which there are 5 books genre
#we will extract the text from these tags one by one by Looping over these tags

all_genre=[]  #empty list for store the
for i in genre[0:5]:
    all_genre.append(i.text.replace('\n',''))
all_genre
```

```python
#lets extract all the tags having the 5 book review
review=soup.find_all('p',class_="excerpt")
review[0:5]
```

```python
#we have all the tags in which there are the reviews of 5 book
#we will extract the text from these tags one by one by Looping over these tags

all_review=[]  #empty list for store the
for i in review[0:5]:
    all_review.append(i.text.replace('\n',''))
all_review
```
